```python
data['name']
data['cut2'] = data['cut2'].apply(lambda x: " ".join(x))
data_featrue = data[data['rating'] >= 4][['name','cut2']]
data_featrue
```

```python
# !pip install sklearn
```

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
```

```python
tran = TfidfVectorizer()
cvt = CountVectorizer(max_features = 500)
```

```python
cv = cvt.fit_transform(_flatten(list(data_featrue['cut2'])))
cv.toarray().shape
```

```python
tfidf = tran.fit_transform(_flatten(list(data_featrue['cut2'])))
tfidf
```

```python
for i in tfidf:
    print(i)
    break
```

```python
from sklearn.model_selection import train_test_split
```

```python
x1,x2,y1,y2 = train_test_split(tfidf,list(data_featrue['name']), test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import LinearSVC
```

```python
model_rf = RandomForestClassifier()
model_rf.fit(x1,y1)
pred_rf = model_rf.predict(x2)
```

```python
model_svc = LinearSVC()
model_svc.fit(x1,y1)
pred_svc = model_svc.predict(x2)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y2,pred_rf))
print(classification_report(y2,pred_svc))
```
